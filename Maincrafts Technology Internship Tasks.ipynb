## Objective
The objective of this project is to perform data cleaning, analysis, and visualization to extract meaningful insights from the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## Loading the Dataset

```python
df_math = pd.read_csv("student-mat.csv", sep=';')
df_por = pd.read_csv("student-por.csv", sep=';')
```

```python
df_math['subject'] = 'Math'
df_por['subject'] = 'Portuguese'

df = pd.concat([df_math, df_por], axis=0)
```

## Understanding the Dataset 
- Checking shape, columns, and data types


```python
df.shape
df.columns
df.info()
df.describe()
```

## Data Cleaning
- Handling missing values
- Removing duplicates

```python
df.isnull().sum()
df = df.dropna()   # or fillna()
df = df.drop_duplicates()
```

## Exploratory Data Analysis (EDA)

### Distribution of Data

```python
sns.histplot(df['G3'], kde=True)
plt.title("Final Grade Distribution")
plt.show()
```

## Data Analysis
Answering key questions based on the dataset.

### 1. Average Final Grade (G3)

```python
avg_grade = df['G3'].mean()
avg_grade
```

### 2. Number of Students Scoring Above 15

```python
high_scores = df[df['G3'] > 15].shape[0]
high_scores
```

### 3. Study Time vs Performance

```python
correlation = df['studytime'].corr(df['G3'])
correlation
```

### 4. Gender-wise Average Performance

```python
gender_avg = df.groupby('sex')['G3'].mean()
gender_avg
```

## Data Visualization

### Histogram of Final Grades

```python
sns.histplot(df['G3'], kde=True)
plt.title("Distribution of Final Grades")
plt.show()
```

### Study Time vs Final Grades

```python
sns.scatterplot(x='studytime', y='G3', data=df)
plt.title("Study Time vs Grades")
plt.show()
```

### Gender vs Average Score

```python
gender_avg.plot(kind='bar')
plt.title("Average Score by Gender")
plt.ylabel("Average G3")
plt.show()
```

## Key Insights

- The average final grade of students is approximately **12**.  
- A total of **122** students scored above 15.  
- There is a **weak positive** correlation between study time and performance.  
- One gender performs slightly better on average based on the dataset.  
- Most students have grades concentrated in the mid-range.
